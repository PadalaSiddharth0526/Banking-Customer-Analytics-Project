## Key EDA Takeaways
- The dataset spans three European markets (France, Germany, Spain) with France representing roughly half the customer base.
- Age is right-skewed, with most customers between 30-50 and a smaller senior population (60+) that, as later analysis shows, churns disproportionately.
- Around a third of customers hold a zero balance, an important segment for targeted engagement.
- Credit scores are approximately normally distributed around the mid-600s, typical of a mainstream retail banking population.
- Most customers hold 1-2 products; very few hold 3-4, representing a cross-sell growth opportunity explored in the segmentation notebook.
